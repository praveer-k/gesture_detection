```python
import mlflow
```

## Metric Upload

Here is a very simple example how to use MLflow, to save a metric to remote MLflow app.

```python
# first we need to set the experiment - we will use set_experiment method to do this, because it will automaticaly
# create experiment if the experiment does not already exist
mlflow.set_experiment('my_first_experiment')
```

```python
# before we upload any objects we need to start a new run
mlflow.start_run()
```

```python
# now, since the experiment has started, we can upload our metric

example_metric = 0.8

mlflow.log_metric('example_metric', value=example_metric)
```

```python
# when we are finished with this experiment we should end the run - it is crucial to do this!
# NOTE: you can resume the run using run_id parameter in the start_run method

mlflow.end_run()
```

## Results

Now log in to your MLflow app and check if the new experiment `my_first_experiment` appeared, where you can display the metric in `run` details. Check the screens below for the comparison.

NOTE: If there is no experiment of such name in the remote MLflow then check the local directory. If you see `mlruns` folder, then it is an indication, that you have not set the environment variables before launching Jupyter Notebook (see the `Usage` section in `README.md` file in this repository <https://github.com/dlabsai/mlflow-for-gcp>).
